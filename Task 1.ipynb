Had 2 options for yearly analysis:

a - create functions and run them on each dataset

b - create a mega dataset with all combined and use group by operations

b seems better to avoid looping through each dataset for each function

```python
#Adding a year clumn so once we combine all datasets we can still perform year wise analysis

df_17["year"]=2017
df_18["year"]=2018
df_19["year"]=2019
df_20["year"]=2020
df_21["year"]=2021
df_22["year"]=2022
df_23["year"]=2023
```

```python
#  Merging the datasets 

dfs = [df_17,df_18,df_19,df_20,df_21,df_22,df_23]
df = pd.concat(dfs,axis=0)
```

Club wise

```python
# Enter a club and see its best performers

def club_perf(club):
    return (
        df[df["Club"] == club][["Name","Overall"]]
        .sort_values(by="Overall", ascending=False)
        .drop_duplicates(subset="Name", keep="first")[["Name", "Overall"]][:10]
    )

club_perf("FC Barcelona")
```

Nation Wise

```python
# Nation wise best preformers

def nation_perf(nation):
    return (
        df[df["Nationality"] == nation][["Name","Overall"]]
        .sort_values(by="Overall", ascending=False)
        .drop_duplicates(subset="Name", keep="first")[["Name", "Overall"]][:10]
    )

nation_perf("France")
```
